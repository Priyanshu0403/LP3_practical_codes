```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')
df
```

```python
print("Description: ")
display(df.describe())
print("\nInfo: ")
display(df.info())
print("\nNull values:")
display(df.isna().sum())
```

```python
# Removing null values
df = df.dropna()
display(df.isna().any())

# Converting numeric columns back to numeric data types
# This is because, since there were null values originally in the dataset, the numeric columns' data types turned into object automatcally
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
from sklearn.preprocessing import StandardScaler
df = df[['QUANTITYORDERED','PRICEEACH','SALES']]

scaler = StandardScaler()
df[['QUANTITYORDERED','PRICEEACH','SALES']] = scaler.fit_transform(df[['QUANTITYORDERED','PRICEEACH','SALES']])
df
```

```python
from sklearn.cluster import KMeans

inertia = []

for i in range(1,11):
    clusters = KMeans(n_clusters=i, init='k-means++', random_state=42)
    clusters.fit(df)
    inertia.append(clusters.inertia_)

plt.figure(figsize=(6,6))
sns.lineplot(x = [i for i in range (1,11)], y=inertia)
plt.grid(True)
plt.show()
```

Q1: What is the primary operational mechanism of the K-Nearest Neighbors algorithm?

A: KNN is a non-parametric, lazy learning algorithm. To classify a new point, it finds the K closest data points (neighbors) in the training set based on a distance metric, typically Euclidean distance. The new point is then assigned the class label that is most frequent among those $K$ neighbors (majority vote).

Q2: Why is feature scaling, such as Standard Scaling, crucial for KNN, and was it performed in this notebook?

A: Feature scaling is critical for KNN because the algorithm is entirely based on distance calculation. Features with larger numerical ranges (like Glucose or Age) will inherently dominate the distance function, making features with smaller ranges appear insignificant, regardless of their actual predictive importance. This notebook did not perform feature scaling, which is a major point of improvement for the model's performance.

Q3: Based on the confusion matrix, explain the meaning of the model's low Recall value ($\approx 0.582$) in a clinical context

.A: Recall is defined as $TP / (TP + FN)$. A low recall means the model has a high rate of False Negatives (FN = 23). In a clinical context, a False Negative is a patient who actually has diabetes but was predicted not to have it. This is a severe clinical error because it results in missed diagnoses and delayed, potentially life-saving, treatment.

Q1: What is the objective function that K-Means attempts to minimize?

A: K-Means seeks to minimize the Inertia, also known as the Within-Cluster Sum of Squares (WCSS). This is the sum of the squared distances between each data point and the centroid of the cluster it belongs to. Minimizing WCSS results in clusters that are tight and internally cohesive.

Q2: What is the Elbow Method, and how do you interpret the plot to find the optimal number of clusters ($K$)?

A: The Elbow Method is a heuristic used to estimate the optimal number of clusters ($K$). It plots the Inertia (WCSS) against the number of clusters ($K$). As $K$ increases, Inertia always decreases. The optimal $K$ is the point on the graph where the rate of decrease in Inertia slows down dramatically, forming an "elbow" shape. The elbow visually marks a good trade-off between minimizing distortion and overfitting.

Q3: The notebook drastically reduced its sample size by dropping all rows with any missing data. What are two alternative strategies for handling null values in a clustering dataset?

A:Imputation: Instead of dropping rows, missing values can be replaced (imputed) with statistical measures, such as the mean, median, or mode of the respective feature column.Specific Feature Removal: If a column (like ADDRESSLINE2 in this data) has too many nulls to be useful, it can be dropped entirely before training the model, preserving all other rows.

Q4: Why is the K-Means algorithm considered a non-deterministic algorithm?

A: K-Means is non-deterministic because the clustering result is dependent on the initial random placement of the cluster centroids. Different initializations can lead to different final cluster assignments and different local minima of the WCSS. The notebook uses init='k-means++' (which helps select better initial seeds) and random_state=42 (to ensure reproducibility of that initial selection).
